```python
import random
import numpy as np
import pandas as pd
```

```python
MATRIX_SIZE = 2000
```

```python
def prost_potok(lmbd):
  time_event = random.expovariate(lmbd)
  return (time_event)
```

```python
time_list = np.zeros([MATRIX_SIZE, MATRIX_SIZE])
lmbd_list = []
```

```python
for i in range(MATRIX_SIZE):
  lmbd = random.uniform(0, 3)
  lmbd_list.append(lmbd)
  counter = 0
  T = 0
  time = 0
  for j in range(MATRIX_SIZE):
    T = prost_potok(lmbd)
    time += T
    time_list[i][j] = time
    counter += 1
```

```python
df = pd.DataFrame(time_list)
df['lmbd_value'] = lmbd_list
```
